```python
import pandas as pd
```

```python
import statsmodels.api as sm
df = pd.read_csv('Advertising.csv')
x = df [['TV','radio','newspaper']]
y = df['sales']
df.head()
```

Find a OLS model with intercept on TV and Radio

```python
x = sm.add_constant(x)
```

```python
x # A new value equaling to 1 has been added as the intercpt value
```

```python
model = sm.OLS(y,x).fit()
```

```python
model.summary()
```

#### R- squared and Adjusted R- squared value almost tending to 1,meaning the model is pretty good

```python
import matplotlib.pyplot as plt
x.iloc[:,1:].corr()
```

Here, the value is almost equal to 0,thus no such multi collinearity found

### Not much relation between the independant features, so no Multi Collinearity issue 

```python
data = pd.read_csv('Salary_Data.csv')
```

```python
x = data[['YearsExperience','Age']]
y = data['Salary']
```

```python
x = sm.add_constant(x)
```

```python
x
```

```python
model = sm.OLS(y,x).fit()
```

```python
model.summary()
```

Check P-value of Age, if we consider 95% covariance getting covered, the value should be less than 0.05, but it is more, thus there can be some Collinearity and we can find it out using corr()
